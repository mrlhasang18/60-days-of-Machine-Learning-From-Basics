# Data Cleaning with Pandas

```python
import pandas as pd

# importing our data set
df = pd.read_excel("Customer Call List.xlsx")
df
```

```python
# Now let's look it's details :
df.info()
```

```python
# Nulls in each column
df.isnull().sum()
```

```python
# Total Nulls in table
df.isnull().sum().sum()
```

# Removing the duplicates using method: drop() in pandas

```python
# Now let's drop all the duplicates values using Drop() method:
df.drop_duplicates()
```

# Let's drop the specific Unnecessary Column using method : drop()

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

# Let's clean the Last Name column with unnecessary characters:

```python
# using strip method for cleaning:
df["Last_Name"] = df["Last_Name"].str.strip("123/._")
df
```

# Cleaning the Phone_Number column using lambda

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"]
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3]+'-'+x[4:7]+'-'+x[8:12])
df["Phone_Number"]
```

```python
# Now let's remove the NaN values using replace method:
df["Phone_Number"] =df["Phone_Number"].str.replace("nan---","")
df["Phone_Number"] =df["Phone_Number"].str.replace("N/a---","")
df["Phone_Number"]
```

```python
df
```

# Cleaning the Address column

```python
# creating a new column 
df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split(',',n=2,expand = True)
df[["Street_Address","State","Zip_Code"]]
```

```python
df
```

# Making Yes and No format consitent:

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer']
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']
```

```python
# Now let's take a look at our table:
df
```

```python
# let's remove this Nan and others, we can use fillna
df.fillna(" ")
```

```python
# LET'S DROP THE ADDRESS
df = df.drop(columns = "Address")
df
```
