```python
import pandas as pd
```

```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000], 'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'], 'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
df  = pd.DataFrame(data = d)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
one = OneHotEncoder(sparse_output = False )
```

```python
ord = OrdinalEncoder()
```

```python
# Importing column transformer
from sklearn.compose import make_column_transformer
```

```python
ct = make_column_transformer(
 (one,['city']),
 (ord,['size']),
  remainder = "passthrough"
)
```

```python
ct.set_output(transform = "pandas")
```

```python
z = ct.fit_transform(df)
```

```python
z
```

```python
# Drop
ct1 = make_column_transformer(
 (one,[1]),
 (ord,[2]),
  remainder = "drop"
)
```

```python
ct1.set_output(transform = "pandas")
```

```python
z = ct1.fit_transform(df)
```

```python
z
```

```python
ct2 = make_column_transformer(
 (one,['city']),
 ('passthrough',['size']),
  remainder = "drop"
)
```

```python
ct2.set_output(transform = "pandas")
```

```python
z = ct2.fit_transform(df)
```

```python
z
```
