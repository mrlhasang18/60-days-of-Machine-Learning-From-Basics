# Seaborn: Visualization tool

```python
import seaborn as sns
# seaborn has built-in datasets
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips
```

# Scatter Plot

```python
sns.scatterplot(y="tip",x="total_bill",data=tips,hue="size",size="size",sizes=(20,200),legend="full")
```

```python
sns.scatterplot(y="tip",x="total_bill",data=tips,hue="day",size="size",sizes=(20,200),legend="full")
```

# Histplot

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

# displot

```python
# similar to histplot
sns.displot(tips['tip'], kde=True, bins=15)
```

# Barplot

```python
tips
```

```python
sns.barplot(x="sex",y="tip",data=tips)
```

# Boxplot and Stripplot

```python
sns.boxplot(x="day",y="tip",data=tips,hue="sex")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex",dodge=True)
```

# Joint plot

```python
# It is combined or merged plot

sns.jointplot(x='tip',y='total_bill',data=tips,kind="reg")
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind="kde",shade=True)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind="hex")
```
